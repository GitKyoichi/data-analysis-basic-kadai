```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)

x = [0,1,2,3]
y = [1,2,-5,2]

# 折れ線グラフ
axes[0, 0].plot(x,y)

# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0,10,100)

y = 2 + 2 * np.sin(2 * x)

# sin関数プロット
axes[0,1].plot(x, y, linewidth=2.0)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# 平均15、標準偏差5の正規分布から2000個の乱数を生成
x = np.random.normal(15, 5, 2000)

# ヒストグラム
axes[0,2].set(xlim=(-10, 40))

axes[0,2].hist(x, bins=30)

np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))

# 散布図
axes[1,0].scatter(x, y)


plt.show()
```
